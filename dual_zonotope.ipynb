```python
import numpy as np

def polar_dual(verts):
    """
    Compute the polar dual of a polytope (here, a zonotope) given its
    vertices, assuming the origin is in the interior.

    P = conv(verts)
    P* = { y : v.y <= 1 for all v in verts }
       = conv(dual vertices)

    The dual vertices are obtained from the FACETS of P:
    each edge (v_i, v_{i+1}) defines a supporting line v.n = c (c>0,
    after orienting normal outward), giving dual vertex n/c.
    """
    verts = np.asarray(verts, dtype=float)
    n = len(verts)

    dual_verts = []
    for i in range(n):
        v1 = verts[i]
        v2 = verts[(i + 1) % n]
        edge = v2 - v1
        # outward normal for CCW-ordered vertices: rotate edge by -90deg
        normal = np.array([edge[1], -edge[0]])
        c = np.dot(normal, v1)  # normal . v = c on this facet
        if c == 0:
            raise ValueError("Facet passes through origin; dual vertex at infinity")
        dual_verts.append(normal / c)

    return np.array(dual_verts)


def polygon_area(pts):
    pts = np.asarray(pts, dtype=float)
    x, y = pts[:, 0], pts[:, 1]
    return 0.5 * abs(np.dot(x, np.roll(y, -1)) - np.dot(y, np.roll(x, -1)))


if __name__ == "__main__":
    #verts = np.array([[3, 3], [1, 3], [-3, -1], [-3, -3], [-1, -3], [3, 1]], dtype=float)
    verts = np.array([[4,3], [2,5], [0,3], [-2,1], [-4,-3], [-2,-5], [0,-3], [2,-1]], dtype=float)

    dual = polar_dual(verts)
    print("Polar dual vertices (scale 1):")
    for v in dual:
        print(f"  ({v[0]:.6f}, {v[1]:.6f})")

    print(f"\nArea of P*  = {polygon_area(dual):.6f}")
    print(f"Area of P*6 = {polygon_area(6 * dual):.6f}")
```
